```python
def compute_dom_prob(k, m, n):
    z = k + m + n
    prob_dom = (
        8 * k * n + 8 * k * m + 4 * m * n + 4 * k**2 - 4 * k + 3 * m**2 - 3 * m
    ) / ((4 * (z**2)) - 4 * z)

    return prob_dom
```

```python
if __name__ == "__main__":
    with open("./datasets/rosalind_iprb.txt", "r") as file:
        k, m, n = [int(i) for i in file.readline().strip().split()]

    prob_dom = compute_dom_prob(k, m, n)
    print(prob_dom)
```
