```python
def compute_gc(sequence):
    c_count = sequence.count("C")
    g_count = sequence.count("G")
    total_count = len(sequence)

    gc_content = ((c_count + g_count) / total_count) * 100

    return gc_content
```

```python
def read_fasta(file_path):
    name_seq = {}

    with open(file_path, "r") as file:
        fasta = file.read()
        records = fasta.lstrip(">").split(">")

    for record in records:
        chunk = record.split("\n")
        name = chunk.pop(0)
        seq = "".join(n for n in chunk if n)

        name_seq.update({name: seq})

    return name_seq
```

```python
if __name__ == "__main__":
    name_seq = read_fasta("./datasets/rosalind_gc.txt")
    name_gc_content = {}

    for name, seq in name_seq.items():
        gc_content = compute_gc(seq)
        name_gc_content.update({name: gc_content})

    ranked_name_gc_content = {
        name: gc_content
        for name, gc_content in sorted(
            name_gc_content.items(), key=lambda item: item[1], reverse=True
        )
    }

    for name, gc_content in ranked_name_gc_content.items():
        print(f"{name}\n{round(gc_content,3)}\n")
```
