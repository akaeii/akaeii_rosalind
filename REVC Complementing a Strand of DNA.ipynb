```python
def complementer(s):
    dna_complement = {"A": "T", "T": "A", "C": "G", "G": "C"}
    rev = s[::-1]
    revc = "".join(dna_complement[b] for b in rev)
    return revc
```

```python
with open("./datasets/rosalind_revc.txt", "r") as file:
    sequence = file.read().strip()
    revc = complementer(sequence)
    print(revc)
```
